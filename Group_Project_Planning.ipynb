## Data Description

For our analysis, we will be using the data scientist job salaries dataset obtained from the website Kaggle.com (https://www.kaggle.com/datasets/ruchi798/data-science-job-salaries/data). The dataset was created using salary data obtained from another website, ai-jobs.net, which collected and aggregated the salary data through survey submissions and jobs that have open salaries. It is then made freely available on the website.

The data scientist job salaries dataset contains 12 columns, with 11 containing variables and 1 serving as a row number column. There are 607 total observations in the dataset with each representing one data science job. The variables in the dataset are listed and can be described as follows:

- `work_year`: A categorical variable that indicates the year that the salary for the job was paid.
- `experience_level`: A categorical variable that indicates the experience level of the job. Contains 4 different levels: 
    - EN (Entry-level or Junior)
    - MI (Mid-level or Intermediate)
    - SE (Senior-level or Expert)
    - EX (Executive-level or Director)
- `employment_type`: A categorical variable that indicates the type of employment of the job. Contains 4 different types:
    - PT (Part-time)
    - FT (Full-time) 
    - CT (Contract) 
    - FL (Freelance)
- `job_title`: A categorical variable that indicates the title of the job.
- `salary`: A numerical variable that indicates the salary paid for the job.
- `salary_currency`: A categorical variable that indicates the type of currency that the salary was paid in. Given in the format of an ISO 4217 currency code.
- `salary_in_usd`: A numerical variable that indicates the the salary for the job in United States dollars (USD). Salaries that are not in USD are converted using the foreign exchange rate divided by the average USD rate for the year through fxdata.foorilla.com.
- `employee_residence`: A categorical variable that indicates the country of residence of the employee working the job. Given in the format of an ISO 3166 country code.
- `remote_ratio`: A categorical variable that indicates how much work for the job was completed remotely. Contains 3 different values:
    - 0 (No remote work, with less than 20% of work being remote)
    - 50 (Partially remote job)
    - 100 (Fully remote work, with more than 80% of work being remote)
- `company_location`: A categorical variable that indicates what country the employer's main office or contracting branch is located in. Given in the format of an ISO 3166 country code.
- `company_size`: A categorical variable that indicates the company's average number of employees over the year. Contains 3 different sizes: 
    - S (small company with less than 50 employees)
    - M (medium company with 50 to 250 employees) 
    - L (large company with more than 250 employees)

## Question

The purpose of this analysis is to answer the question: **What characteristics of a data science job and/or data science employee are associated with a data scientist's salary in USD?** Determining what characteristics about a job and employees are associated can be useful, as it can inform data scientists who are looking for jobs on what usually is associated with jobs with higher salaries. It can also be used in future analyses to compare data science jobs with other jobs in technology sector.

To address this question, we will use the data scientist job salaries dataset to construct a model by using `salary_in_usd` as the response variable and the other variables as potential input variables. The input variables used in the model will be selected from the dataset through a selection process that will be covered in the **Methods and Plan** section. Since, the dataset contains variables such as experience level, company_size, etc., which are characteristics of a data science job and/or data science employee, the variables in the dataset are relevant and will be useful in solving the question of interest. Building a model determining if these variables have an association with `salary_in_usd` would allow us to obtain estimates, interpret the results, and draw conclusions about the population. This, in turn, will help us identify some characteristics of a data science job and/or data science employee that would answer the question.

The question of interest is focused on inference, as the main goal of this analysis is to develop an understanding of the relationship between the response and input variables. We would answer the question by estimating what occurs in the population using a linear regression fitted to a sample, and we would also draw conclusions about the population using said linear regression. These two aspects of the analysis follow what would be done in statistical inference.

## Methods and Plan

### The Model

To answer the question, we will build an additive multiple linear regression (MLR) with salary in USD as the response variable. The input variables we will include in the model will be all of the variables in the dataset, except for the variables `salary` and `salary_currency`. The analysis will not include `salary` as it is essentially a non-transformed version of the `salary_in_usd` (response variable). Including it as an input variable would be redundant as they measure similar things, and it would be difficult to use as a response as the salaries between currencies are not comparable due to not being in the same units. The type of currency the salary is in is not included due to the response already being all in USD, the currencies being hard to merge into small enough groups effectively analyze, and the regions where the employer and employee are located providing similar information. One of the variables `company_location` or `employee_residence` will also not likely be included in the model, as the exploratory analysis showed that both variables are fairly similar.

We are using a MLR due to the response variable being continuous (the data is in integers; however, salaries can have decimal values) and wanting to include multiple input variables in the model. We are using an additive model, which will not include any interaction terms. This is because, to answer the question of interest, we are only interested in determining the association between individual input variables and the response.

The assumptions for using a linear regression model that will be tested once the model is fit include the data having a linear relation, the errors being independent, the conditional distribution of the error terms following a Normal distribution, having constant variance of error terms, and no multicollinearity. The linear assumption would be difficult to show visually; however, the linear model does work on categorical variable and acts like a two-sample t-test, so the assumption can be fulfilled.

### The Plan

We will use backward, stepwise selection to select the input variables that will be used in the MLR. This process involves starting with a full model with all of the input variables and then removing the least useful variables one by one from the model. The least useful variable is selected in each step by creating models that each exclude only one of the variables of the model found in the previous step. The model with the highest $R^2$ value is then selected and determined to be the best model of that step (the one without the least useful variable). We would then have all of the best models for each step and size, and we would compare their adjusted $R^2$ values to determine the best overall model obtained through the backward stepwise selection process.

Using this process for model selection allows us to efficiently and systematically select a good model for the data. It allows for a model to be selected without fitting as many models, making it less computationally demanding. The issue with using this method for model selection is that it might not produce the best possible model for the data. This is due to the backward selection process evaluating fewer models and certain variable combinations that could have led to a better model being skipped over, a result of only removing variables and not considering if re-adding them can improve the model. Another potential issue would be that the number of samples must be larger than the number of variables in order to fit a full model. However, this requirement will be a non-issue for this analysis as the sample size is large enough. 

Before the backward selection is performed, the sample will be randomly split into two datasets, with one containing 30% of the data being used as the model selection set (undergoes backward selection) and the other 70% of the data being used as a training set (to build and evaluate the model). This is done because using the same data to select the variables and estimate the model will violate the assumptions of the analysis and will invalidate the results of the statistical inference. The variables would fit the data well as they are selected because they fit the data. We want to ensure that the model not only effectively models the specific sample but can be just as effective when its variables are fitted to a new sample. This is so we know the variables selected for the model are not a result of the data obtained in the specific sample and can be applied to other samples as well, which in turn would better work to estimate the unknown population.

We will then use the training set to fit a model containing the input variables determined by the best model and test the assumptions of the linear regression model. If the model is adequate, we can then use hypothesis testing to determine what characteristics of a data science job and/or data science employee are associated with a data scientist's salary in USD.

2. Perform backward selection on the selection set.
3. Choose the best model determined by backward selection.
4. Use the training set to fit a model containing the input variables determined by the best model.
5. Test the model to ensure it statisfies its assumptions.
6. Analyze the model and interpret results.
